```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt

# Combine all datasets into a list
datasets = [DS1, DS2, DS3, DS4, DS5, DS6, DS7, DS8, DS9, DS10]

# Extract the 'U' variable for a specific location and level from each dataset
u_data = [
    ds['U'].sel(latitude=20, longitude=200, level=14, method='nearest')[:32]
    for ds in datasets
]

# Combine into an xarray.Dataset with an ensemble dimension
u_ensemble = xr.concat(u_data, dim='ensemble')

# Apply rolling mean with a 4-day window and minimum periods of 1
u_ensemble = u_ensemble.rolling(time=5, min_periods=1).mean()

# Compute the ensemble mean
u_mean = u_ensemble.mean(dim='ensemble')

# Save the dataset with the ensemble dimension
output_file = "ensemble_u_data.nc"
# # u_ensemble.to_netcdf(output_file)

# u_ensemble = xr.open_dataset('./ensemble_u_data.nc')['U']

# Plot the data
plt.figure(figsize=(18, 6))
colors = plt.cm.viridis(np.linspace(0, 1, len(u_data)))  # Use a consistent colormap

# Plot individual ensemble members
for i, u in enumerate(u_ensemble):
    u.plot(color='k', alpha=0.7)

# Plot the ensemble mean
u_mean.plot(color='red', linewidth=2)

# Add legend and labels
plt.title('U - Latitude 20N, Longitude 200E', fontsize=15)
plt.xlabel('Time',fontsize=17)
plt.ylabel('U [m/s]',fontsize=17)
plt.grid(True)
plt.tight_layout()
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
# Save the plot
plot_file = "ensemble_u_plot.png"
plt.savefig(plot_file, dpi=300, bbox_inches='tight')

plt.show()

print(f"Ensemble dataset saved to: {output_file}")
print(f"Plot saved to: {plot_file}")
```
